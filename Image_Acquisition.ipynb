```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/drive/MyDrive/Colab Notebooks/orange.png')
cv2_imshow(img)
print(img.shape)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_resize = cv2.resize(img,[250,250])
cv2_imshow(img_resize)
```

```python
img_flip = cv2.flip(img,0) #Flip with respect to x-axis
img_flip1 = cv2.flip(img,1) #Flip with respect to y-axis
img_final = cv2.hconcat([img_flip,img_flip1])
cv2_imshow(img_final)
```
